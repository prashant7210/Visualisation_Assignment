### 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
feature_1 = np.random.normal(0, 1, 100)
feature_2 = 0.5 * feature_1 + np.random.normal(0, 1, 100)
feature_3 = -0.2 * feature_1 + np.random.normal(0, 1, 100)
feature_4 = 0.3 * feature_1 + 0.5 * feature_2 + np.random.normal(0, 1, 100)
feature_5 = np.random.normal(0, 1, 100)

data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4,
    'Feature 5': feature_5
})
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm')
```

## Plotly Assignment

### 1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.

```python
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = df['X'], y = df['Y'], z = df['Z'], mode = 'markers', marker_size = 4))
```

### 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

```python
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
fig = px.density_heatmap(df, x = 'Month', y = 'Day', z = 'Sales', color_continuous_scale = 'Viridis')
fig.show()
```

### 4. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin((x + y))

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
fig = go.Figure(data=[go.Surface(
    z=df['Z'].values.reshape(x.shape),
    x=df['X'].values.reshape(x.shape),
    y=df['Y'].values.reshape(y.shape),
    colorscale='Viridis'
)])
fig.show()
```

## BOKEH Assignment

### 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
p = figure(title = 'Sine Wave', x_axis_label = 'x', y_axis_label = 'sin(x)')
p.line(x, y, line_width = 2)
show(p)
```

### 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
x = np.random.rand(100)*100
y = np.random.rand(100)*100
sizes = np.random.rand(100)*50 + 10
colors = [(int(r), int(g), int(b)) for r, g, b in np.random.randint(0, 256, (100, 3))]
p = figure(title = 'Scatter Plot',
           x_axis_label = 'x - axis',
           y_axis_label = 'y - axis')
p.scatter(x, y, size = 0.25*sizes, color = colors)
show(p)
```

### 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
##### fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
##### counts = [20, 25, 30, 35]


```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
p = figure(x_range = fruits, title = 'Fruit Counts', x_axis_label = 'Fruits', y_axis_label = 'Counts')
p.vbar(x = fruits, top = counts, width = 0.5)
show(p)
```

### 4. Create a Bokeh histogram to visualize the distribution of the given data.
##### data_hist = np.random.randn(1000)
##### hist, edges = np.histogram(data_hist, bins=30)

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
p = figure(title = 'Histogram', x_axis_label = 'Value', y_axis_label = 'Frequency')
p.quad(top = hist, bottom = 0, left = edges[:-1], right = edges[1:], line_color = 'black')
show(p)
```

### 5. Create a Bokeh heatmap using the provided dataset.
##### data_heatmap = np.random.rand(10, 10)
#### x = np.linspace(0, 1, 10)
#### y = np.linspace(0, 1, 10)
#### xx, yy = np.meshgrid(x, y)

```python
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
p = figure(title = 'Heatmap', x_axis_label = 'x', y_axis_label = 'y')
p.image(image = [data_heatmap], x = [xx.min()], y = [yy.min()], dw = [xx.max() - xx.min()], dh = [yy.max() - yy.min()])
show(p)
```
